```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the predicted dataset
csv_path = r"C:\Users\sagni\Downloads\Insider Threat Detection\threat_users.csv"
df = pd.read_csv(csv_path)

# Display prediction counts
print("ðŸ”¢ Predicted Threat Label Counts:\n")
print(df['predicted_label'].value_counts())

# Visualize prediction distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='predicted_label', palette='Set2')
plt.title('Distribution of Predicted Insider Threats')
plt.xlabel('Threat Label')
plt.ylabel('User Count')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

# Filter high-risk users (Malicious + Compromised)
high_risk = df[df['predicted_label'].isin(['Malicious Insider', 'Compromised Account'])]

# Preview top high-risk users
print("\nðŸš¨ Sample High-Risk Users:")
print(high_risk.head())

# Save high-risk users to a new file
output_path = r"C:\Users\sagni\Downloads\Insider Threat Detection\high_risk_users.csv"
high_risk.to_csv(output_path, index=False)
print(f"\nâœ… High-risk users saved to: {output_path}")
```
